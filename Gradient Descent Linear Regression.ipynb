```python
lr = 0.0001
Epochs = 1000
n = float(len(x))
theta_0 = 20
theta_1 = 0
```

```python
def derivative_theta_0(x, y, y_p, n):
    return (-1/n) * sum(y - y_p)
```

```python
def derivative_theta_1(x, y, y_p, n):
    return (-1/n) * sum((y - y_p) * x)
```

```python
store_theta_0 = []
store_theta_1 = []

for i in range(Epochs):
    y_p = theta_0 + theta_1 * x
    theta_0 = theta_0 - lr * derivative_theta_0(x, y, y_p, n)
    theta_1 = theta_1 - lr * derivative_theta_1(x, y, y_p, n)
    store_theta_0.append(theta_0)
    store_theta_1.append(theta_1)
    
print(theta_0)
print(theta_1)
```

```python
y_pred = theta_0 + theta_1 * x

plt.figure()
plt.scatter(x, y)
plt.plot([min(x), max(x)], [min(y_pred), max(y_pred)], color='r')
plt.show()
```
