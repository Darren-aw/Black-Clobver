```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('weight-height.csv')
x = df.iloc[:, 2]
y = df.iloc[:, 1]
n = len(y)
```

```python
x_normalized = (x - x.mean())/x.std()
x_normalized
```

```python
plt.plot(x_normalized, y, '.')
```

```python
X = np.c_[np.ones(n), x_normalized]
X
```

```python
θ = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
θ
```

```python
#xc = x_normalized[np.argsort(x_normalized)]
plt.scatter(x, y, color='#ef5423', label='Scatter')
plt.plot(x, θ[0] + θ[1]*x_normalized, color='#58b970', label='Regression Line')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.legend()
plt.show()
```
